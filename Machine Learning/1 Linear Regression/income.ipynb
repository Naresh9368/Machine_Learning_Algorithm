```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df['income']=df['per capita income (US$)']
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='yellow',marker='*')
```

```python
df=df.drop('per capita income (US$)',axis='columns')
df
```

```python
new_df=df.drop('income',axis='columns')
```

```python
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,df.income)
```

```python
model.predict([[2020]])
```
