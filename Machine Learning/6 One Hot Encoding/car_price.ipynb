```python
import pandas as pd
```

```python
df=pd.read_excel('carprices.xlsx')
df
```

```python
df['car_model']=df['Car Model']
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.scatter(df.Mileage,df.Sell_Price,marker='+',color='red')
```

```python
%matplotlib inline
plt.scatter(df['Age(yrs)'],df.Sell_Price,marker='+',color='red')
```

```python
df.drop('Car Model',axis='columns')
```

```python
dummies=pd.get_dummies(df.car_model)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop('car_model',axis='columns')
```

```python
final
```

```python
final=final.drop('Car Model',axis='columns')
```

```python
final
```

```python
final=final.drop('Mercedez Benz C class',axis='columns')
final
```

```python
x=final.drop('Sell_Price',axis='columns')
x
```

```python
y=final.Sell_Price
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
x_test
```

```python
model.predict([[35000,3,0,1]])
```
