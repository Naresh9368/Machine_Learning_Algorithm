```python
dummies_1=pd.get_dummies(df.Department)
dummies_2=pd.get_dummies(df.salary)
```

```python
dummies_1.head()
```

```python
dummies_2.head()
```

```python
merged=pd.concat([df,dummies_1,dummies_2],axis='columns')
merged.head()
```

```python
final=merged.drop(['Department','salary'],axis='columns')
final
```

```python
final=final.drop(['technical','medium'],axis='columns')
final.head()
```

```python
x=final.drop('left',axis='columns')
x.head()
```

```python
y=final['left']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```
