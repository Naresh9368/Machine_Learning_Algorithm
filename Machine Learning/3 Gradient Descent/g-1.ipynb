```python
def gd(x,y):
    m_curr = b_curr = 0
    n=len(x)
    learning_rate=0.08
    for i in range(100):
        yp=m_curr*x+b_curr
        cost=(1/n)*sum([var**2 for var in (y-yp)])
        md=-(2/n)*sum(-x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print("m {}, b {}, itration {}".format(m_curr,b_curr,i))
```
